## Data Pre-Processing

```python
import re
def casefolding(Comment):
    Comment = Comment.lower()
    Comment = Comment.strip(" ")
    Comment = re.sub(r'[?|$|.|!²_:")(-+,]','',Comment)
    Comment = re.sub(r"\d+", "", Comment)
    Comment = Comment.translate(str.maketrans("","",string.punctuation))
    Comment = Comment.strip()
    return Comment
```

```python
#proses tokenizing

def token(Comment):
    nstr = Comment.split(' ')
    dat= []
    a = -1
    for hu in nstr:
        a = a + 1
    if hu == '':
        dat.append(a)
    p = 0
    b = 0
    for q in dat:
        b = q - p
        del nstr[b]
        p = p + 1
    return nstr
```

```python
#Proses Filtering
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
stopwords_ind = stopwords.words ('indonesian')

from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

more_stopword = ['<USERNAME>', 'yg', 'jg', 'sj', 'username', 'sm', 'nya','jd', 'tp', 'zzzzz', 'aja', 'tuh', 'amin', 'ya', 'ga', 'gak', 'mah']
stopwords_ind = stopwords_ind + more_stopword

def remove_stop_words (text):
    clean_words = []
    text = text.split()
    for word in text: 
        if word not in stopwords_ind:
            clean_words.append(word)
    return " ".join(clean_words)
```

```python
def text_preprocessing_process(Comment):
    text = casefolding(Comment)
    text = token(Comment)
    text = remove_stop_words(Comment)
    text = stemming_sastrawi(text, stemmer)
    return text
```

## Feature Extraction

```python
X = data['Comment']
y = data['Sentiment']
```

```python
#proses TF-IDF

import pickle
from sklearn.feature_extraction.text import TfidfVectorizer
vec_TF_IDF = TfidfVectorizer(ngram_range=(1,1))
vec_TF_IDF.fit(X)

X_tf_idf = vec_TF_IDF.transform(X)

pickle.dump(vec_TF_IDF.vocabulary_,open("feature_tf-idf.sav", "wb"))
```

## Feature Selection (Chi Square)
